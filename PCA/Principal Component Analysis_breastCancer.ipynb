```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
data_raw = load_breast_cancer()
data_raw.keys()
```

```python
df = pd.DataFrame(data_raw['data'], columns = data_raw['feature_names'])
df.head()
```

```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
df_scaler = scaler.transform(df)
```

```python
#Perform the principal component analysis transformation
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(df_scaler)

```

```python
x_pca = pca.transform(df_scaler)

print(x_pca.shape)
print(df.shape)
```

```python
#Visualize 2 dimensions principal components
plt.scatter(x_pca[:,0],x_pca[:,1], c=data_raw['target'], cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#Investigating at the principal components
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=data_raw['feature_names'])
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
